```python
"""Part 3: Unpacking Product Ratings
Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

Measures of Centrality:

Calculate the mean, median, and mode for the rating of products.
How do customers generally rate products? Is there a common trend?"""
```

```python

# Calculate the mean rating
mean_rating = df['stars'].mean()
print(f"Mean Rating: {mean_rating:.2f}")

# Calculate the median rating
median_rating = df['stars'].median()
print(f"Median Rating: {median_rating}")

# Calculate the mode of the ratings
mode_rating = df['stars'].mode()
print(f"Mode Rating: {mode_rating.values}")
```

```python
"""Measures of Dispersion:

Determine the variance, standard deviation, and interquartile range for product rating.
Are the ratings consistent, or is there a wide variation in customer feedback?"""
```

```python
# Calculate the variance of the ratings
variance_rating = df['stars'].var()
print(f"Variance: {variance_rating:.2f}")

# Calculate the standard deviation of the ratings
std_deviation_rating = df['stars'].std()
print(f"Standard Deviation: {std_deviation_rating:.2f}")

# Calculate the interquartile range (IQR) of the ratings
Q1 = df['stars'].quantile(0.25)
Q3 = df['stars'].quantile(0.75)
iqr_rating = Q3 - Q1
print(f"Interquartile Range (IQR): {iqr_rating:.2f}")
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Calculate skewness of the ratings
rating_skewness = skew(df['stars'])
print(f"Skewness: {rating_skewness:.2f}")

# Calculate kurtosis of the ratings
rating_kurtosis = kurtosis(df['stars'], fisher=False)  # Use `fisher=False` for Pearson's definition
print(f"Kurtosis: {rating_kurtosis:.2f}")
```

```python
#Positive Skewness: Indicates a distribution with a tail on the right (i.e., more lower ratings, few higher).

#Kurtosis < 3: Indicates a platykurtic distribution; flatter with lighter tails.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['stars'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Q-Q Plot
plt.figure(figsize=(10, 5))
stats.probplot(df['stars'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Rating')
plt.show()
```
